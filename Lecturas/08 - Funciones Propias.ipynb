# Funciones propias

En la mayoría de los casos, resolver un problema de programación no se limita a buscar una función en un módulo que pueda llamar y que resuelva el problema completo. Es por ello que además de poder usar funciones que alguien más definió, Python permite **definir nuestras propias funciones**.

Las funciones son particularmente útiles para encapsular partes de una solución, para luego combinarlas en la solución general. En la práctica, el usar funciones es la mejor estrategia para **reducir la complejidad** del código que estamos construyendo.

Para comprender de mejor manera cómo se definen las funciones propias, se trabajará pensando en que se debe realizar una función que calcule la sumatoria desde 0 hasta un determinado número entero:

## Estructura de Funciones Propias

### Encabezado

Para crear una función se debe comenzar con la palabra reservada **`def`**. Esta palabra debe ir acompañada por el nombre que utilizaremos para invocar la función y sus parámetros de entrada. Todo esto corresponde al **encabezado** de la función.

El nombre de la función debe ser representativo a lo que se está realizando y debe partir con un verbo en infinitivo; si el nombre contiene 2 o más palabras, estas deben ir unidas mediante un guión bajo.

A continuación del nombre de la función, se colocan los paréntesis, dentro de los cuales se ingresa el (o los) parámetro(s) de entrada de la función. En otras palabras, los datos que se necesitan para que la función pueda realizar el proceso. En caso de que se necesiten más de un parámetro de entrada, estos van separados por coma. A estas entradas se les llama **parámetros formales** y toman este nombre debido a que solo indican el nombre genérico de la o las entradas requeridas al momento de definir o formalizar la función.

Finalmente, el encabezado termina con el signo de **dos puntos**, dando pie a que comience a definirse el cuerpo de la función.

Con lo anterior, el **encabezado** de la función queda expresado de la siguiente manera:

```python
def calcular_sumatoria(numero):
```

### Cuerpo

Luego del encabezado, se define el cuerpo de la función, que corresponde al bloque de sentencias que deben ejecutarse al momento de evaluar la función. En otras palabras, el proceso que queremos realizar. 

Cabe señalar que el cuerpo de la función debe escribirse indentado respecto al encabezado de la función, es decir, tiene una sangría mayor desde el margen izquierdo. Se debe tener en cuenta que todas las sentencias de un mismo bloque deben tener la misma indentación.

Siguiendo con el ejemplo anterior, el **cuerpo** del cálculo de una sumatoria viene dado de la siguiente manera:

```python
       i = 0
       suma = 0

       while i <= numero:
              suma = suma + i
              i = i + 1
```

### Retorno de la función

Para dar término a la función creada y entregar su resultado (o salida), se utiliza la palabra reservada **`return`**. Esta palabra reservada provoca que la función finalice y devuelva el valor calculado.

Para finalizar y marcar el fin de la declaración de la función creada, se deja al menos una línea en blanco sin indentar. Entre cada par de funciones definidas y el código que viene antes o después de la definición, se deben dejar dos líneas en blanco.

En la práctica, la palabra reservada `return` es una orden que indica a Python que ha alcanzado un punto en donde **debe** terminar la ejecución de la función.

Una función puede tener varios `return` en el código, pero una vez que el código alcance uno de ellos, la función terminará, entregando el valor que se encuentre acompañando a la palabra reservada `return`. Para que un código funcione de manera correcta, se recomienda utilizar sentencias de decisión en donde cada camino tenga su propia salida (`return`)

Una confusión habitual entre los/as usuarios/as se produce entra la función `print()` y la palabra reservada `return`. La primera de ellas entrega un mensaje o valor que se desea visualizar (mensaje de salida por pantalla), en cambio el `return` genera un valor o tipo de dato que se puede manipular según sea necesario.

Siguiendo con el ejemplo anterior, el uso de la palabra reservada `return` en el cálculo de una sumatoria viene dado de la siguiente manera:

```python
      i = 0
      suma = 0

      while i <= numero:
            suma = suma + i
            i = i + 1
      return suma
```

Con lo anteriormente descrito, ya se pueden unir las partes y formar la función propia creada.

````python
def calcular_sumatoria(numero):
  i = 0
  suma = 0
  
  while i <= numero:
    suma = suma + numero
    i = i + 1 

  return suma
````

Como la finalidad de crear funciones es poder resolver problemas repetitivos o disminuir los cálculos realizados, es fundamental que se *documente* la función, es decir, indicar que es lo que hace. En general una buena documentación de una función debería tener como mínimo:

*   Qué es lo que hace la función.
*   Cuál o cuáles son las entradas que la función requiere para funcionar correctamente, indicando el tipo de cada dato.
*   Cuál es la salida que genera la funcióm, indicando tipo de dato.

Si además usamos un string multilínea (o *docstring*) directamente bajo el encabezado, Python mostrará el texto cuando se desee usar la función.

````python
def calcular_sumatoria(numero):
  """Función que realiza la sumatoria hasta un número dado

  Entrada: un número entero (int)
  Salida: un número entero (int)
  """
  i = 0
  suma = 0
  while i <= numero:
    suma = suma + numero
    i = i + 1 
  return suma
````

Los comentarios que se agregaron al ejemplo anterior se consideran **obligatorios** a efectos de buenas prácticas de programación para todas las funciones. Lo anterior se debe a que los programadores y otros usuarios deben entender cuál es el proceso que la función realiza sin tener que consultar el detalle del código.

## Estructura de Programas

Ahora que ya se conoce como crear funciones propias, se puede inferir que la ubicación de éstas en un programa no es arbitraria. Esto se debe a que los programas poseen una estructura que ayuda a la legilibilidad y que estos puedan ser leídos y entendidos fácilmente por otros programadores o usuarios.

Para organizar los programas, se utilizará una estructura básica que consta de 2 grandes clasificaciones o bloques:
*   Bloque de Definiciones
*   Bloque Principal

### Bloque de Definiciones

En este bloque se realiza la definición de las herramientas requeridas para que el programa pueda ejecutarse. Normalmente este bloque debería simplemente *cargar* herramientas a la memoria y no debería realizar procesos como pedir datos al usuario o realizar impresiones por pantalla.

En el Bloque de Definiciones, también puede sub-dividirse en:

* **Importación de módulos/funciones**: las primeras líneas indican los módulos, o funciones de módulos, que el programa requiere para que el código funciones.

* **Definición de Constantes**: se definen las constantes, las cuales son valores que **no** se alteran durante todo el código. Las cuáles van con nombres capitalizados (Por ejemplo: PI, ARCHIVO_ENTRADA, etc.)

* **Definición de Funciones**: se indican las funciones que el programador ha creado para este programa en particular.

### Bloque Principal

En este bloque se operan los contenidos o funciones que se declararon en el bloque anterior. Es en este apartado donde se trabajará con valores reales para que las funciones puedan realizar los cálculos requeridos, es decir, se *invoca* a las funciones definidas para que realicen las instrucciones indicadas en el cuerpo de cada una de ellas, lo que equivale a evaluar cada función con valores dados. A su vez, se realiza la interacción entre todas las variables definidas (cuál se ejecuta primero, cuál depende de una o de otra...).



El Bloque Principal se divide en tres partes, que hemos venido usando durante el curso:
* **Entrada**: corresponde al punto del código en donde se *alimenta* con datos externos al programa. Hasta el momento hemos visto que esto se hace con la función nativa `input()`, sin embargo, podríamos definir funciones propias para leer datos desde un archivo o solicitar los datos y validar que son correctos.
* **Procesamiento**: corresponde a las transformaciones, cálculos o procesos procesos que se hacen sobre las entradas.
* **Salida**: corresponde a la parte del código encargada de hacer entrega de los resultados requeridos por los/as usuarios/as. Hasta el momento hemos usado `print()` pero existen otras opciones para informar al usuario.

Utilizando el ejemplo anterior, que se solicita crear una sumatoria para un número dado, se realizará un programa completo considerando los Bloque anteriormente mencionados:

````python
### BLOQUE DE DEFINICIONES

## Importación de módulos/funciones
# No Aplica

## Definición de Constantes
# No Aplica

## Definición de Funciones
def calcular_sumatoria(numero):
  """Función que realiza la sumatoria hasta un número dado

  Entrada: un número entero (int)
  Salida: un número entero (int)
  """
  i = 0
  suma = 0

  while i <= numero:
    suma = suma + i
    i = i + 1 

  return suma


### BLOQUE PRINCIPAL
## Entrada
numero_solicitado = input("Favor ingresar número para calcular sumatoria: ")

## Procesamiento
numero_solicitado = int(numero_solicitado)
sumatoria = calcular_sumatoria(numero_solicitado)

## Salida
print("La sumatoria desde el número 1 al número", numero_solicitado, "es", sumatoria)
````

## Parámetros Formales y Actuales

Cuando se define una función, se declaran parámetros de entrada, los cuales reciben el nombre de **parámetros formales** y son las que se utilizan para las operaciones que se encuentran en el cuerpo de dichas funciones.

Estos parámetros son definidos con nombres genéricos y Python los utiliza para determinar la cantidad de entradas que requieren las funciones y para saber los distintos procesos o transformaciones que sufren determinadas entradas.

A modo de ejemplificar lo anterior, se utiliza la ecuación física que define la obtención del Trabajo Mecánico (**T**), a partir de algunos parámeros dados:

$$T = F\cdot d\cdot \cos(\theta)$$

En este caso, los datos de entrada o parámetros formales son $F$ (fuerza), $d$ ( desplazamiento) y $\theta$ (ángulo que se forma).

Por otro lado, cuando se requiere realizar el cálculo o evaluar una función, se requieren datos concretos, ya sea solicitados al usuario, extraídos de algún documento o de alguna plataforma. A este tipo de dato, que es concreto, se le denomina **parámetros actuales**

Utilizando el ejemplo del Trabajo Mecánico, para poder realizar el cálculo de esta variable se deben conocer los valores de $F$, $d$ y $x$.

Si:

$$F = 10 N$$

$$d = 5 m$$

$$\theta = \frac{\pi}{2}$$

Con los valores anteriores, se puede indicar que el Trabajo mecánico es:

$$T = 10\cdot 5\cdot \cos\left(\frac{\pi}{2}\right) = 50 J$$

Los valores 10 N, 5 m y $\frac{\pi}{2}$ son considerados **parámetros actuales**.


En conclusión:

*   **Parámetros Formales**: son los nombres que se le asignan a los datos de entrada al momento de **definir** las funciones.
*   **Parámetros Actuales**: son los valores que se le asignan a los datos de entrada al momento de **evaluar** las funciones.



## Alcance (*Scope*) 

Al definir funciones propias en el código, encapsulamos secciones de código dentro de una función. Esto implica que lo que sea que definamos dentro de una función **solo existe dentro de ella**. A este concepto se le conoce como *scope*.

Se definen dos *scopes* distintos en el programa, es decir, dos entornos en el que existen valores. Cómo para acceder a estos valores normalmente usamos variables, se identifica el *scope* de ellas hablando de **variables globales** y **variables locales**.

*   **Variables Globales**: son aquellas a las que puedo acceder en todo el programa, es decir, puedo modificarlas tanto dentro de una función como fuera de ellas.
*   **Variables Locales**: son aquellas variables que solo existen en el contexto de una función, por lo que no puedo acceder a ellas fuera del código y una vez que la función alcance el **`return`**, estas desaparecen.

A continuación se ilustra el concepto de *scope*. Consideremos la función `es_primo(n)` y el programa que sigue.

````python
def es_primo(n):
  ''' Función que determina si un entero positivo es primo o no.
      Entrada: n, entero positivo mayor que 2,
      Salida: True si es primo, False en caso contrario.
  '''
  i = 2
  while i < n:
    if n % i == 0:
      return False
    i = i + 1
  return True

valor = input('Ingrese un entero positivo: ')

valor = int(valor)

resultado = es_primo(valor)

if resultado:
  print('El número', valor, 'es primo')
else:
  print('El número', valor, 'no es primo')
````

## Funciones para simplificar la vida

Finalmente, las funciones son una poderosa herramienta para simplificar nuestros programas, pues permiten transformar procesos complejos, en programas más simples.

Consideremos el siguiente ejercicio:

Construya una función en Python que calcule la exponenciación entre dos números enteros positivos usando solamente operadores de suma y resta, considere para ello que:

$3^4 = 3 * 3 * 3 * 3 = 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 + 3 $ 

Este problema en particular es complejo, pues requiere varios ciclos anidados para resolverse. La solución en una única función quedaría como:

````python
def exponenciacion(base, exponente):
  i = 1
  potencia = 0
  aux = base
  while i < exponente:
    j = 0
    multiplicacion = 0
    while j < aux:  
      k = 0
      suma = 0
      while k < base:
        suma = suma + 1
        k = k + 1
      multiplicacion = multiplicacion + suma
      j = j + 1
    potencia = multiplicacion
    aux = multiplicacion
    i = i + 1
  return potencia
  
print(exponenciacion(3,4))
````

Sin embargo, podríamos definir *funciones auxiliares* y facilitar el entendimiento del problema.

Aunque esto alarga un poco el código, hace que sea mucho más legible y fácil de entender.

En este caso, la función `multiplicacion(a, b)` define la multiplicación como un conjunto de sumas, lo que es un problema trivial. Cómo este proceso es sencillo y ya logré aislarlo en una función, luego puedo usarlo en la función exponenciación para ir calculando:
$1 * 3 = 3$

$3 * 3 = 9$

$9 * 3 = 27$

$27 * 3 = 81$

Qué corresponde al proceso iterativo de la potencia. Lo cuál es mucho más amigable que los tres ciclos anidados que hay en el caso anterior.

````python
def multiplicacion(a, b):
  i = 0
  resultado = 0
  while i < b:
    resultado = a + resultado
    i = i + 1
  return resultado

def exponenciacion(base, exponente):
  i = 0
  potencia = 1
  while i < exponente:
    potencia = multiplicacion(potencia, base)
    i = i + 1
  return potencia

print(exponenciacion(3,4))
````
