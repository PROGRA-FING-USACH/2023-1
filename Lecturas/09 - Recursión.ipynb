## Definición

La recursión es un concepto fundamental en la programación y en las ciencias de la computación. Es una técnica en la que una función se llama a sí misma, directa o indirectamente, para resolver un problema dividiéndolo en subproblemas más pequeños. La recursión puede ser útil para resolver problemas de manera eficiente y elegante en muchos casos.

En el ejemplo anterior, el proceso de "*ir a la casa X, golpear la puerta y ver si vive nuestro amigo*" es un proceso que se repitió hasta lograr el objetivo, que es "*encontrar a nuestro amigo*"


Veamos un ejemplo más práctico que se puede encontrar en la vida diaria:

matrioskas3.jpg

Las muñezas Matrioskas son un excelente ejemplo para comprender el concepto de **recursividad**.

Cuando se abre la primera muñeca, aparece una nueva muñeca
Luego, se abre la siguiente muñeca y aparece una nueva...

Estos mismos pasos se repiten hasta que se abra una muñeca y en su interior no hay nada.

En otras palabras, el proceso es "abrir muñeca" y se repite hasta que no exista muñeca alguna.

Si se lleva el concepto de recursividad a térmimos académicos, ésta suele aparecer de manera recurrente en el área de las matemáticas:

A modo de ejemplo se tiene cuando se quiere definir un número factorial:

n! = n * (n-1)!

Esta definición de número factorial se puede encontrar explicado de la siguiente manera:

Rec.JPG

Al observar el ejemplo anterior, el cual corresponde a la definición matemática de un número factorial, se puede concluir que el paso 2 se repetirá hasta que el valor de **n** alcance el valor de 1 o 0.

Lo anterior se conoce como un proceso **recursivo**.

Considerando todo lo anterior, en el ámbito de la programación, una función es considerada recursiva cuando ésta se **llama a si misma**, de manera directa o indirecta.

La finalidad de usar esta metodología es resolver un problema dado a través de la resolución de instancias más pequeñas del problema.

Para crear una función recursiva, se debe considerar que debe existir:
* Una **regla de recursión**, vale decir, una regla que llama a la misma función, pero con parámtros distintos.
* Un **condición de borde**, o caso base, que detendrá el proceso.

En el caso de un número factorial:

* **Regla de recursión**: n*(n - 1)!, si n > 1
* **Condición de borde**: 1, si n = 1 o n= 0

Una vez definidos la regla de recursión y condición de borde, ya se puede comenzar a programar:

```python
### BLOQUE DE DEFINICIONES

# Definición de Funciones
def calcular_factorial_r(n):
  """
  Fución que calcula el factorial de un número de manera recursiva
  Entrada: un número (int)
  Salida: un número (int)
  """
  if n > 1:
    # Regla de recursión
    return n * calcular_factorial_r(n - 1)
  else:
    #Condición de borde
    return 1

### BLOQUE PRINCIPAL

## Entrada
numero = input("Favor ingresar un número entero positivo: ")

## Procesamiento
numero = int(numero)
factorial_r = calcular_factorial_r(numero)

## Salida
print("El factorial de", numero, "es", factorial_r)
```

El ejemplo anterior muestra el cálculo de una función recursiva para el cálculo del factorial de un número. A continuación, se muestra otra opción para calcular el factorial de un número, pero esta vez, de manera iterativa.

Se dejan ambas alternativas para que vean las similitudes y diferencias en las formas de desarrollar una solución.

```python
### BLOQUE DE DEFINICIONES

# Definición de Funciones
def calcular_factorial_i(n):
  """
  Fución que calcula el factorial de un número de manera iterativa
  Entrada: un número (int)
  Salida: un número (int)
  """
  i = 1
  factorial = 1
  while i <= n:
    factorial = factorial * i
    i = i + 1
  return factorial


### BLOQUE PRINCIPAL
## Entrada
numero = input("Favor ingresar un número entero positivo: ")

## Procesamiento
numero = int(numero)
factorial_i = calcular_factorial_i(numero)

## Salida
print("El factorial de", numero, "es", factorial_i)
```

Se debe tener en cuenta que al realizar una función recursiva, en la regla de recursión el valor del parámetro actual (valores de entrada de la función) **deben cambiar** en cada llamado. En caso de no hacerlo, **nunca se alcanzará la condición de borde**.

Una función puede ser resuelta de forma **recursiva** o de forma **iterativa**. La solución será más simple o compleja, más fácil o difícil de implementar dependiendo de cada caso. 


# Recomendaciones y precauciones al usar recursión

**Condición de borde o caso base:** Asegúrate de tener una condición de borde o caso base bien definida en tu función recursiva. Esta condición es necesaria para detener la recursión y evitar un ciclo infinito que agote los recursos del sistema.

**Modificar los parámetros en cada llamada recursiva:** Es fundamental cambiar los valores de los parámetros en cada llamado recursivo para que la función progrese hacia el caso base. Si no modificas los parámetros, la función puede quedar atrapada en un bucle infinito y causar errores.

**Profundidad de recursión:** Ten en cuenta que Python tiene un límite en la profundidad de recursión para evitar el agotamiento de recursos en caso de un bucle infinito. Si tu función recursiva supera este límite, Python lanzará una excepción RecursionError

**Elegir la mejor solución:** La recursión no siempre es la mejor solución para un problema. A veces, las soluciones iterativas pueden ser más eficientes y fáciles de entender o viceversa. Compara las soluciones recursivas e iterativas y elige la que mejor se adapte a tus necesidades y requisitos de rendimiento.

**Pruebas y depuración:** Asegúrate de probar exhaustivamente tu función recursiva con diferentes entradas para garantizar que funcione correctamente. La depuración de funciones recursivas puede ser más complicada que la de funciones iterativas, por lo que es fundamental entender cómo funciona la recursión en tu función y cómo se relaciona con el caso base.
